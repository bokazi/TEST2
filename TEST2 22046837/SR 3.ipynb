```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv(r'social-development_zaf.csv')

# 1. Bar chart for Country_Name vs Value
plt.figure(figsize=(10, 6))
sns.barplot(x='Country_Name', y='Value', data=data)
plt.title('Value by Country')
plt.xlabel('Country Name')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()

# 2. Bar chart for Country_ISO3 vs Value
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Value', data=data)
plt.title('Value by Year')
plt.xlabel('Year')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()

# 3. Bar chart for Indicator Name vs Value
plt.figure(figsize=(10, 6))
sns.barplot(x='Indicator Name', y='Value', data=data)
plt.title('Value by Indicator Name')
plt.xlabel('Indicator Name')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Value', data=data, estimator=sum)  # Using sum as estimator for bar graph
plt.title('Value by Year')
plt.xlabel('Year')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()

```
